## Cleaning and Combining Files

### Cleaning Info

```python
info.info()
```

```python
#Checking for duplicates
info.duplicated().sum()
```

```python
#Checking for missing values
info.isna().sum()
```

```python
# Splitting Hero|Publisher into two columns
info[["Hero","Publisher"]] = info["Hero|Publisher"].str.split('|',expand=True)
info.head()
```

```python
# Dropping Original Column
info = info.drop(columns=["Hero|Publisher"])
info.head()
```

```python
#Checking type of Measurements column
measurements = info.loc[0,"Measurements"]
print(type(measurements))
measurements
```

```python
## use .str.replace to replace all single quotes
info["Measurements"] = info["Measurements"].str.replace("'",'"')
## Apply the json.loads to the full column
info["Measurements"] = info["Measurements"].apply(json.loads)
info["Measurements"].head()
```

```python
#Checking type of Measurements column
measurement = info.loc[0,"Measurements"]
print(type(measurement))
measurement
```

```python
height_weight = info["Measurements"].apply(pd.Series)
height_weight
```

```python
# concat long_lat with original dataframe
info = pd.concat((info, height_weight), axis = 1)
info.head(2)
```

```python
# Dropping Original Column
info = info.drop(columns=["Measurements"])
info.head()
```

```python
info[["Height (cm)", "units"]] = info["Height"].str.split(" ", expand = True)
info[["Weight (kg)", "unit"]] = info["Weight"].str.split(" ", expand = True)
```

```python
info["Height (cm)"] = info["Height (cm)"].astype(float)
```

```python
info["Weight (kg)"] = info["Weight (kg)"].astype(float)
```

```python
info = info.drop(columns=["unit", "units"])
```

```python
info.head()
```

### Cleaning Powers

```python
powers.info()
```

```python
powers.head()
```

```python
#Checking for duplicates
powers.duplicated().sum()
```

```python
#Checking for missing values
powers.isna().sum()
```

```python
#Checking type of Measurements column
power = powers.loc[0,"Powers"]
print(type(power))
power
```

```python
#Splitting string into list based off of comma separated values
powers["Powers"] = powers["Powers"].str.split(",")
powers.head()
```

```python
power = powers.loc[0, "Powers"]
print(type(power))
power
```

```python
powers.head()
```

```python
#Turning each cell in each row into a string of a list
for i in range (len(powers)):
    newCell = str(powers.loc[i, "Powers"]).replace("'", '"')
    powers.loc[i, "Powers"] = newCell

# Apply json.loads to entire column
powers["Powers"] = powers["Powers"].apply(json.loads)
powers.head()
```

```python
## exploding the column of lists
exploded = powers.explode("Powers")
exploded.head(5)
```

```python
## saving the unique values from the exploded column
cols_to_make = exploded["Powers"].dropna().unique()
```

```python
#Warning! This calls a Pink Wall of Doom (non-urgent) if you run this with
# warnings on. Code is functional. 
for col in cols_to_make:
    powers[col] = powers["Powers"].str.contains(col, regex = False)
powers.head()
```

### Combining Files

```python
info.head()
```

```python
powers.head()
```

```python
new_df = pd.merge(info, powers, left_on = "Hero", right_on = "hero_names")
new_df.head()
```

## Answering Questions

### Compare the average weight of super powers who have Super Speed to those who do not.

```python
sns.barplot(data = new_df, x = "Super Speed", y = "Weight (kg)")

plt.title("Average weight dependent on Super Speed")
plt.xlabel("Has Super Speed")
plt.ylabel("Weight (in kg)")
plt.grid(axis = "y")
```

We can see here that superheroes that have super speed, tend to weigh more on average than those that do not.

 ### What is the average height of heroes for each publisher?

```python
sns.barplot(data = new_df, x = "Publisher", y = "Height (cm)")
plt.title("Average Height of superhero based off of Publishing company")
plt.xlabel("Publisher")
plt.ylabel("Hero Height in cm")
plt.grid(axis = "y")
plt.xticks(rotation = 90)
```

We can see here that all heroes published by Image Comics are on average taller than 200 cm.
